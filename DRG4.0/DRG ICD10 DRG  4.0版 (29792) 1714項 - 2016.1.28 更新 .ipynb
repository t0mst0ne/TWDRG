# 讀取 DRG 4.0版 1700項 權重表 

```python
import pandas as pd
import numpy as np

```

# 讀取DRG 4.0 版 分類表

```python
#sheet = pd.read_excel('NHI/29792_1_Tw-DRGs分類表4.0版ICD-10_1050128.xls', sheetname=[ x for x in range(4,27)], index_col=None)
sheet = pd.read_excel('NHI/29792_1_Tw-DRGs分類表4.0版ICD-10_1050128.xls', sheetname=[ i for j in (range(4,27), range(33, 34)) for i in j], index_col=None)
```

## 把sheet [4:27] 列表 7.1  讀進到 dataframe

```python
sheetname=[ i for j in (range(4,27), range(33, 34)) for i in j]
#for tab in range(4,27):
for tab in sheetname:
    DRG = sheet[tab]
    DRG = DRG.reset_index()
    print DRG.head(1)
    step1 = DRG [ pd.notnull(DRG[DRG.columns[0]]) & pd.isnull(DRG[DRG.columns[1]]) & pd.notnull(DRG[DRG.columns[2]]) & pd.notnull(DRG[DRG.columns[3]]) ]
    step1 = step1[['index',DRG.columns[2], 'Unnamed: 2']]
    step1.columns = ['DRG','CM-PCS','en']
    step1.loc[:,'DRG'] = step1.loc[:,'DRG'].str.replace(u'\u3001',',' )
    for x in step1.DRG.unique():
        #print x
        y = x.split(',')
        #print y
        for z in y:
            #print z
            #step3 = step1 [ step1.DRG.str.contains(x) ]
            step3 = step1 [ step1.DRG.str.contains(x).fillna(False) ]
            step3.DRG = z
            step3.loc[:,'DRG'] = step3.loc[:,'DRG'].replace(r'^(\d\d\d)$','00\\1' , regex=True)
            step3.loc[:,'DRG'] = step3.loc[:,'DRG'].replace(r'^(\d\d\d\d)$','0\\1' , regex=True)
            step3.loc[:,'DRG'] = step3.loc[:,'DRG'].replace(r'^(\d+)','DRG\\1' , regex=True)
            #print step3.head(1)
            step3.to_csv('DRG4/%s.csv' %(step3.DRG.unique().tolist()[0]), index=False, encoding='utf-8')
    step2 = DRG [ DRG[ DRG.columns[1] ].str.contains('DRG|N') & pd.notnull(DRG[DRG.columns[2]]) & pd.isnull(DRG[DRG.columns[3]]) ]
    step2[step2.columns[[1,2]]].to_csv('DRG4/tab_%s.csv' %(tab),index=False, encoding='utf-8',sep='\t',header=False)
    
```

# 讀取 DRG 4.0版 1714項 權重表 - 權重表還沒出來

```python
df0 = pd.read_excel('NHI/30143_1_105年4.0版1716項Tw-DRGs適用權重表_1050128.xls', header = 14 )
cost1 = df0[:-1] 
cost1.loc[:,'DRG'] = cost1.loc[:,'DRG'].replace(r'^(\d\d\d)$','00\\1' , regex=True)
cost1.loc[:,'DRG'] = cost1.loc[:,'DRG'].replace(r'^(\d\d\d\d)$','0\\1' , regex=True)
cost1.loc[:,'DRG'] = cost1.loc[:,'DRG'].replace(r'^(\d+)','DRG\\1' , regex=True)
print len(cost1)
cost1
```

# Merge DRG cost and PCS
* cat DRG* >> Total.csv  ;  cat N* >> Total.csv
* cat tab* >> name_DRG.4.0.csv 

```python
DRG3 = pd.read_csv('DRG4/Total.csv')
DRG_name = pd.read_csv('DRG4/name_DRG.4.0.csv',sep='\t', header=None)
DRG_name.columns = ['DRG','DRG-name']
# Merge DRG and DRG name
f1 = pd.merge(DRG_name,DRG3,on='DRG',how='right')
# Merge DRG and cost
final = pd.merge(f1,cost1,on='DRG',how='right')  

final.columns = ['DRG','DRG-NAME','CM-PCS','CM-PCS-Name','MDC','流水號','RW','個案數<20註記','邏輯註記','幾何平均住院日','下限臨界點','上限臨界點']
final['定額'] = final['RW']*39277*(1+0.061)
#final [ final.DRG.str.contains(r'^N01') ]
final.to_csv('DRG4/final-4.0.1.30.csv' , index=False,encoding='utf-8')
```
